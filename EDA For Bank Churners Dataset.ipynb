## EDA For Bank Churners Dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("F:\Data Analytics For Coding\Python\Python EDA\BankChurners.csv")
data
```

## Which customer_Age has the highest Attrited Customer?

```python
attrited_customers = data.loc[data["Attrition_Flag"] == "Attrited Customer"]
attrited_customers
```

```python
attrition_count = attrited_customers.groupby("Customer_Age").size().reset_index(name = "count").sort_values("count",ascending = False).iloc[0]
attrition_count
```

```python
attrited_customers["Customer_Age"].value_counts().reset_index().iloc[0]
```

# What is the distribution of Credit_Limit by Income_Category?

```python
plt.figure(figsize = (10,5))
sns.kdeplot(data = data, x = "Credit_Limit", hue = "Income_Category")
plt.title("Distribution of Credit_Limit by Income_Category")
plt.show()
```

# Which has the higher attrition rate among certain Education_Level or Marital_Status groups?

```python
plt.figure(figsize = (10,5))
sns.barplot(data = data, x = "Education_Level", y = "Customer_Age", hue = "Marital_Status")
plt.title("Attrition rate among certain Education_Level or Marital_Status groups")
plt.show()
```

# Is there any correlation between Marital_Status and Attrition_Flag?

```python
plt.figure(figsize = (10,5))
sns.barplot(data = data, x = "Marital_Status", y = "Attrition_Flag")
plt.gca().invert_yaxis()
plt.show()
```
